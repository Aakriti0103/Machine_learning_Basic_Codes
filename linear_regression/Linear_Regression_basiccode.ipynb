```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
%matplotlib inline
plt.scatter(df.year,df['per capita income (US$)'],color='r',marker='+')
```

```python
#From the above plot we can see it is linear regression
reg=linear_model.LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df[['year']],df['per capita income (US$)'],test_size=0.1)
```

```python
reg.fit(x_train,y_train)
reg.predict(x_test)
reg.score(x_test,y_test)
```
